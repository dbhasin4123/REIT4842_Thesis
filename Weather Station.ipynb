```python
import numpy as np

def find_closest_bureau(df):
    result_df = pd.DataFrame()
    
    # Extract the basic information
    result_df['ID'] = df.iloc[:, 0]
    result_df['Latitude'] = df.iloc[:, 1]
    result_df['Longitude'] = df.iloc[:, 2]
    result_df['State'] = df.iloc[:, 3]
    
    # Get distances and generators names (all columns from column 5 onwards)
    distance_df = df.iloc[:, 4:]
    
    # Find the minimum distance and corresponding bureau for each row
    result_df['Generator'] = distance_df.idxmin(axis=1)
    result_df['Distance'] = distance_df.min(axis=1)
    
    return result_df
```

```python
result = find_closest_bureau(df)

# Display the results
print(result)
```

```python
import pandas as pd
import numpy as np

def find_closest_ids(df):
    """
    Analyze a dataframe to find the closest ID for each generator based on distance.
    
    Parameters:
    df (pandas.DataFrame): DataFrame with first 4 columns as 'ID', 'latitude', 'longitude', 'state',
                          and remaining columns are generator names with distance values
    
    Returns:
    pandas.DataFrame: A table with generator names, closest ID, and the distance
    """
    result_df = pd.DataFrame()
    
    # Extract the basic information
    result_df['ID'] = df.iloc[:, 0]
    result_df['Latitude'] = df.iloc[:, 1]
    result_df['Longitude'] = df.iloc[:, 2]
    result_df['State'] = df.iloc[:, 3]
    
    # Get distances and generators names (all columns from column 5 onwards)
    distance_df = df.iloc[:, 4:]
    
    # Create empty lists to store results
    generators = []
    closest_ids = []
    min_distances = []
    closest_latitudes = []
    closest_longitudes = []
    closest_states = []
    
    # For each generator column
    for generator in distance_df.columns:
        # Find the index of the minimum distance for this generator
        min_idx = distance_df[generator].idxmin()
        
        # Get the minimum distance
        min_distance = distance_df.loc[min_idx, generator]
        
        # Add to our lists
        generators.append(generator)
        closest_ids.append(df.iloc[min_idx, 0])  # ID
        min_distances.append(min_distance)
        closest_latitudes.append(df.iloc[min_idx, 1])  # Latitude
        closest_longitudes.append(df.iloc[min_idx, 2])  # Longitude
        closest_states.append(df.iloc[min_idx, 3])  # State
    
    # Create the final result dataframe
    final_result = pd.DataFrame({
        'Generator': generators,
        'Closest_ID': closest_ids,
        'Distance': min_distances,
        'Latitude': closest_latitudes,
        'Longitude': closest_longitudes,
        'State': closest_states
    })
    
    return final_result

# Example usage:
# Assuming df is your dataframe with the first 4 columns as ID, latitude, longitude, state
# and the rest of the columns are generator names with distance values
result = find_closest_ids(df)
print(result)
```

```python
solar_duids = gen_df.loc[gen_df["Fuel Source - Primary"] == "Solar", "DUID"].tolist()
print(solar_duids)
```

```python
def check_solar_duids_in_results(solar_duids, result_df):
    """
    Check if each name in solar_duids list is present in the results dataframe.
    
    Parameters:
    solar_duids (list): List of solar generator names
    result_df (pandas.DataFrame): Result from find_closest_ids function
    
    Returns:
    pandas.DataFrame: DataFrame showing whether each solar DUID is in the results
    """
    # Create a dictionary to store results
    results = {
        'Solar_DUID': [],
        'In_Results': [],
        'Closest_ID': [],
        'Distance': []
    }
    
    # Check each solar DUID
    for duid in solar_duids:
        # Check if this DUID is in the Generator column of result_df
        if duid in result_df['Generator'].values:
            # Get the row for this generator
            row = result_df[result_df['Generator'] == duid].iloc[0]
            
            results['Solar_DUID'].append(duid)
            results['In_Results'].append(True)
            results['Closest_ID'].append(row['Closest_ID'])
            results['Distance'].append(row['Distance'])
        else:
            results['Solar_DUID'].append(duid)
            results['In_Results'].append(False)
            results['Closest_ID'].append(None)
            results['Distance'].append(None)
    
    # Create DataFrame from results
    check_df = pd.DataFrame(results)
    
    # Summary statistics
    total_duids = len(solar_duids)
    duids_in_results = check_df['In_Results'].sum()
    duids_not_in_results = total_duids - duids_in_results
    
    print(f"Total solar DUIDs: {total_duids}")
    print(f"DUIDs found in results: {duids_in_results} ({duids_in_results/total_duids*100:.1f}%)")
    print(f"DUIDs not found in results: {duids_not_in_results} ({duids_not_in_results/total_duids*100:.1f}%)")
    
    return check_df

# Example usage in your notebook:
# Assuming you've already run find_closest_ids and have the result in 'result_df'
check_result = check_solar_duids_in_results(solar_duids, result)
display(check_result)
```
